```python
#SSH Keys - another way
with open('bruce_rsa.pub', 'r') as file:
    raw = file.read()

pkey = base64.b64decode(raw.split()[1])

parts = []
while pkey:
    length = int.from_bytes(pkey[:4], 'big')
    parts.append(int.from_bytes(pkey[4:4+length], 'big'))
    pkey = pkey[4+length:]

parts
```
